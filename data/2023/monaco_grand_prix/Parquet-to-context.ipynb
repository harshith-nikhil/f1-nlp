```python
data = pd.read_csv("qualifying_results.csv")
data
```

```python
def build_context(dridictionary, team, qualifying_position, Q1_time, Q2_time, Q3_time):
    template = f'''
Driver - {dridictionary}
Team - {team}
Qualifying Position - {qualifying_position}
Laptimes - Q1 {Q1_time}, Q2 {Q2_time}, Q3 {Q3_time} 


    '''
    
    return template

merge_context = []
for row in data.iterrows():
    dictionary = dict(row[1])
    template = build_context(dictionary['driver'], 
                             dictionary['team'], 
                             dictionary['qualifying_position'], 
                             dictionary['Q1_time'],dictionary['Q2_time'],dictionary['Q3_time'])
    merge_context.append(template)

print(" ".join(merge_context))
```
